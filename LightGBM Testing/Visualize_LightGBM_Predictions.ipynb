```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Load the submission file
file_path = "submission.csv"  # Update if your file is elsewhere
df = pd.read_csv(file_path)
df.head()
```

```
# Display column names for verification
df.columns
```

```
# Plot the distribution of predicted probabilities
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(df["has_graduated_prob"], bins=50, kde=True, color="steelblue")
plt.title("Distribution of Predicted Graduation Probabilities (LightGBM)", fontsize=14)
plt.xlabel("Predicted Probability of Graduation")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
